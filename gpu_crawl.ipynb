```python
from requests_html import HTMLSession, HTML
import requests
import pandas as pd
from selenium import webdriver
```

```python
#Khi chạy nhớ locate vị Chromedriver
browser = webdriver.Chrome(executable_path='chromedriver.exe')
browser.get('https://www.videocardbenchmark.net/gpu_list.php')
html_text = HTML(html=browser.page_source)
```

```python
#Crawl gpu
df = pd.DataFrame([], columns = ['Name', 'Mark']) 
center = html_text.find('center')
a = center[0].find('tr')
for i in range(1,len(a)):
    s = a[i].text.split('\n')
    name = s[0]
    mark = s[1]
    print(i, name, mark)
    df2 = pd.DataFrame([[name, mark]], columns = ['Name', 'Mark'])
    df = df.append(df2, ignore_index=True)
```

```python
#Chuyển điểm Passmark về dạng int64 để thuận lợi tính toán
df = df[df['Mark'] != 'Passmark G3D Mark']
df = df.reset_index(drop = True)
df['Mark'].astype('int64')
```

```python
#Lưu dữ liệu gpu
df.to_csv(path_or_buf='data_gpu.csv', index=False, sep='\t')
```

```python
#Đọc dữ liệu gpu
df = pd.read_csv('data_gpu.csv', sep='\t')
```

```python
#Tạo field Keyword chứa keyword từ tên của GPU
df['Keyword'] = [-1]*len(df)
```

```python
#Keyword sẽ là tên GPU nhưng được cách ra và chuyển về dạng viết thường
for i in range(len(df)):
    df['Keyword'][i] = df['Name'][i].lower().split()
```

```python
df = df[['Keyword', 'Mark']]
df.to_csv(path_or_buf='data_gpu.csv', index=False, sep='\t')
df.head()
```
